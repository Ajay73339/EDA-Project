# Introduction

The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset contains around 40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate.

## Summary of the dataset



UID - ID

Id's -
( ID, CollegeID, CollegeCityID)

continuous -
(Salary, 10percentage, 10board, 12percentage, CollegeGPA, English, Logical, Quant,ComputerProgramming, ElectronicsAndSemicon, ComputerScience, MechanicalEngg, ElectricalEngg, TelecomEngg, CivilEngg )

continuous/Standardized -
(Domain, conscientiousness, agreeableness, extraversion, neuroticism, openess_to_experience)

Categorical -
(Designation, JobCity, Gender, CollegeTier, Degree, Specialization, CollegeCityTier, CollegeState, 12board)

Date -
(DOJ, DOL, DOB, 12graduation, GraduationYear)

```python
# getting the dataset columns
data.columns
```

```python
data.info()
```

From above we can see that some datetime columns are not set to date time so, first we set the date time columns to their respective DataTypes.

```python
today = pd.to_datetime('today').strftime('%m/%d/%Y')
data.loc[data['DOL'] == 'present', 'DOL'] = today
```

```python
date_col_list = ['DOJ', 'DOL', 'DOB', '12graduation', 'GraduationYear']
```

```python
for i in range(len(date_col_list)):
  data[date_col_list[i]] = pd.to_datetime(data[date_col_list[i]])
```

```python
data = data[data['Salary'] < 800000]
data = data[data['Logical'] < 780]
data = data[data['12percentage'] > 40]
data = data[data['English'] > 200 ]
data = data[data['ElectronicsAndSemicon'] < 600 ]
data = data[data['extraversion'] > -3]
data = data[data['agreeableness'] > -3]
data = data[data['openess_to_experience'] > -4]
data = data[data['collegeGPA'] > 20]
```

```python
data.head()
```

```python
sns.stripplot(x='Gender', y='Salary', data=data)
plt.title('Stripplot')
plt.show()
```

The above graph is the depiction of the salary earned based on gender.

Observations:
- on average both male and female earn around 3-4 lakhs.
- least salary for both the genders is around same.
- compared to females, males are earning high salaries.

```python
sns.violinplot(x='Degree', y='Salary', data=data)
plt.title('violinplot')
plt.show()
```

Graph depicts the salaries earned based on the degree.

Observations:
- M.Tech./M.E. has high average salary compared to other degrees. Followed by M.Sc.(Tech),B.Tech/B.E. and MCA respectively.
- Highest salary is earned by students who completed M.Tech Degree.
- Bulk of the graduates who pursed MCA are earning below 2lakhs

```python
sns.pointplot(x='Degree', y='12percentage', data=data)
plt.title('pointplot')
plt.show()
```

Plot between marks in 12th to the Degree they chose.

Obsevations:
- Most who got high marks in 12th chose to pursue b.Tech/B.E. and M.Tech/M.E.
- Least marks scored students choose to pursue M.Sc

```python
sns.set(style="whitegrid")

cross_tab = pd.crosstab(data['Degree'], data['Gender'])

cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Degree Vs Gender')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Plot between Degree Chosen to the Gender

Observations:
- most from both Gender prefer B.Tech Degree
- Almost negligible from both the Gender prefer to pursue M.Sc
- Females are less to pursue M.Tech when compared to males.

```python
sns.set(style="whitegrid")

cross_tab = pd.crosstab(data['Specialization'] == 'computer science & engineering', data['Gender'])

cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('CSE vs Gender')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
